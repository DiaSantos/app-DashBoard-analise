```python
#alteração de nomes dos atributos
senioridade ={
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executive'
}
df['senioridade'] = df['senioridade'].replace(senioridade)
df['senioridade'].value_counts()
```

```python
contrato ={
    'FT': 'Tempo_integral',
    'CT': 'Contrato',
    'PT': 'Tempo_parcial',
    'FL': 'Freelance'
}
df['contrato'] = df['contrato'].replace(contrato)
df['contrato'].value_counts()
```

```python
tamanho_empresa ={
    'S': 'Pequena',
    'M': 'Media',
    'L': 'Grande',

}
df['tamanho_empresa'] = df['tamanho_empresa'].replace(tamanho_empresa)
df['tamanho_empresa'].value_counts()
```

```python
remoto ={
    0: 'Presencial',
    50: 'Hibrido',
    100: 'Remoto',
}
df['remoto'] = df['remoto'].replace(remoto)
df['remoto'].value_counts()
```

##Analise Gráfica

```python
df_limpos.head()
```

```python
df_limpos['senioridade'].value_counts().plot(kind='bar', title='Distribuição de senioridade');
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
ordem = df_limpos.groupby('senioridade')['usd'].mean().sort_values(ascending=False).index#criando variavel para calcular média
plt.figure(figsize=(10,6));#Tamanho do gráfico
sns.barplot(data = df_limpos, x = 'senioridade', y ='usd', order=ordem);#dados a serem plotados
plt.title('Salário médio por senioridade');#Título do gráfico
plt.xlabel('Senioridade');#Título do eixo X
plt.ylabel('Salário médio anual(USD)');#Título do eixo Y
plt.show();#Plota o gráfico
```

```python
plt.figure(figsize=(10,4))
sns.histplot(df_limpos['usd'], bins= 50 , kde=True)
plt.title('Distribuição dos salários anuais')
plt.xlabel('Salário em USD')
plt.ylabel('Frequência')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x= df_limpos['usd'])
plt.title('')
plt.xlabel('Salário em USD')
plt.show()
```

```python
ordem_senioridade = ['Junior','Pleno', 'Senior', 'Executive']
plt.figure(figsize=(8,4))
sns.boxplot(x='senioridade', y='usd', data=df_limpos, order=ordem_senioridade, palette='Set2', hue='senioridade')
plt.title('Distribuição salárial por nível de experiência')
plt.xlabel('Nível de Experiência')
plt.ylabel('Salário em USD')
plt.show()
```

```python
# Calcular a média salarial por nível de senioridade
avg_salary_by_seniority = df_limpos.groupby('senioridade')['usd'].mean().reset_index()

# Definir a ordem dos níveis de senioridade
ordem_senioridade = ['Junior', 'Pleno', 'Senior', 'Executive']

# Criar o gráfico de barras usando Plotly Express
fig = px.bar(
    avg_salary_by_seniority,
    x='senioridade',
    y='usd',
    category_orders={'senioridade': ordem_senioridade},
    title='Média Salarial por Senioridade (USD)',
    labels={'senioridade': 'Nível de Experiência', 'usd': 'Média Salarial Anual (USD)'}
)

# Ajustar o tamanho do gráfico
fig.update_layout(width=600, height=400)

fig.show()
```

```python
remoto_contagem = df_limpos['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
             names = 'tipo_trabalho',
             values = 'quantidade',
             title= 'Proporção dos tipos de trabalho',
             hole=0.5 # cria o tipo rosca

)
fig.update_traces(textinfo='percent+label')
fig.show()
```

```python
# Filtrar o DataFrame para o cargo de Data Scientist
df_data_scientist = df_limpos[df_limpos['cargo'] == 'Data Scientist']

# Calcular a média salarial por país
avg_salary_by_country = df_data_scientist.groupby('residencia')['usd'].mean().reset_index()

# Ordenar os países pela média salarial em ordem decrescente
avg_salary_by_country = avg_salary_by_country.sort_values(by='usd', ascending=False)

# Criar o gráfico de barras usando Plotly Express
fig = px.bar(
    avg_salary_by_country,
    x='residencia',
    y='usd',
    title='Média Salarial para Data Scientists por País (USD)',
    labels={'residencia': 'País de Residência', 'usd': 'Média Salarial Anual (USD)'}
)

fig.show()
```

```python
#Criação de dataframepara teste e substituição de valores
import numpy as np

df_salarios = pd.DataFrame({
    'nome': ['Anna', 'Bruno','Carlos','Danielle','Dan'],
    'salario':[4000, np.nan, 5000, np.nan, 100000 ],
})
#substitui o NAN pela média
df_salarios['salario_media']=df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

#substitui o NAN pela mediana
df_salarios['salario_mediana']=df_salarios['salario'].fillna(df_salarios['salario'].median())
df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    'dia_semana': ['Segunda', 'Terca', 'Quarta', 'Quinta', 'Sexta'],
    'temperatura':[30, np.nan, np.nan, 28, 27],
})
#substitui pelo valor anterior
df_temperaturas['preenchi_ffill']= df_temperaturas['temperatura'].ffill()
df_temperaturas
```

```python
df_temperaturas = pd.DataFrame({
    'dia_semana': ['Segunda', 'Terca', 'Quarta', 'Quinta', 'Sexta'],
    'temperatura':[30, np.nan, np.nan, 28, 27],
})
#substitui pelo valor posterios
df_temperaturas['preenchi_bfill']= df_temperaturas['temperatura'].bfill()
df_temperaturas
```
